1. time

Tipo: Fecha/hora

Qué significa: Momento exacto en que se midió la información.

Por qué importa: Permite ver cómo cambian las métricas con el tiempo (gráficas de evolución).

2. site

Tipo: Texto

Qué significa: Nombre del lugar o centro (por ejemplo, Madrid, Valencia).

Por qué importa: Permite comparar el rendimiento entre diferentes centros federados.

3. cell_id

Tipo: Texto

Qué significa: Identificador único de la “celda” o antena 5G.

Por qué importa: Cada celda cubre una zona distinta; te ayuda a saber cuál rinde mejor o peor.

4. user_id

Tipo: Texto

Qué significa: Identificador del usuario o dispositivo conectado (móvil, sensor, PC, etc.).

Por qué importa: Permite saber si un problema afecta a un usuario concreto o a todos.

5. slice_id

Tipo: Texto

Qué significa: Tipo de “slice” o servicio 5G (por ejemplo: eMBB, URLLC, IoT).

Por qué importa: En 5G se dividen recursos según el tipo de aplicación (vídeo, IoT, baja latencia, etc.).

6. rsrp_dbm

Tipo: Número (dBm)

Qué significa: Potencia de señal recibida por el usuario. Cuanto más negativo → peor.

Por qué importa: Indica la calidad de cobertura. Por debajo de -100 dBm = señal débil.

7. rsrq_db

Tipo: Número (dB)

Qué significa: Calidad de la señal recibida, considerando ruido e interferencias.

Por qué importa: Detecta interferencias o saturación en el canal de radio.

8. sinr_db

Tipo: Número (dB)

Qué significa: Relación señal/ruido: mide la calidad general del enlace.

Por qué importa: Cuanto más alto, mejor comunicación y menos errores.

9. throughput_dl_mbps

Tipo: Número (Mbps)

Qué significa: Velocidad de descarga (del servidor al usuario).

Por qué importa: Indica el rendimiento percibido por el usuario al recibir datos.

10. throughput_ul_mbps

Tipo: Número (Mbps)

Qué significa: Velocidad de subida (del usuario al servidor).

Por qué importa: Mide qué tan rápido puede enviar datos (vídeos, sensores, etc.).

11. latency_ms

Tipo: Número (milisegundos)

Qué significa: Tiempo que tarda un paquete en ir y volver (retardo).

Por qué importa: Si es alto (>30 ms), la conexión se siente lenta.

12. jitter_ms

Tipo: Número (milisegundos)

Qué significa: Variación del retardo (inestabilidad en los tiempos de transmisión).

Por qué importa: Si es alto, hay cortes en vídeo o voz.

13. packet_loss_pct

Tipo: Porcentaje (%)

Qué significa: Porcentaje de paquetes de datos perdidos durante la transmisión.

Por qué importa: Si supera el 2 %, hay cortes o fallos perceptibles.

14. prb_utilization_pct

Tipo: Porcentaje (%)

Qué significa: Uso de los bloques físicos de recursos (PRB) de la antena.

Por qué importa: Indica si la celda está congestionada o tiene capacidad libre.

15. ho_success_rate

Tipo: Porcentaje (%)

Qué significa: Tasa de éxito de los “handovers” (cuando un usuario cambia de celda).

Por qué importa: Si baja, los usuarios pierden conexión al moverse.

16. energy_watts

Tipo: Número (W)

Qué significa: Energía consumida por la celda o equipo.

Por qué importa: Permite analizar eficiencia energética o sostenibilidad del sistema.

17. alarm_rate

Tipo: Porcentaje (%)

Qué significa: Porcentaje de alarmas activas o eventos de fallo en el sistema.

Por qué importa: Refleja el estado general de salud de la red.

```python
import pandas as pd
import numpy as np
import datetime

np.random.seed(42)
start = datetime.datetime(2025, 11, 7, 10, 0, 0)
minutes = 60 * 12  # 12 horas de datos

# Fija una semilla aleatoria (seed) para que los resultados sean reproducibles.
# Define un inicio de simulación (10:00 AM).
# Genera datos para 12 horas (720 minutos).

sites = ['madrid_centro', 'valencia_fp']
cells = ['cell_01', 'cell_02', 'cell_03']
users = [f'UE_{i:03d}' for i in range(1, 31)]
slices = ['eMBB', 'URLLC', 'IoT']

# Simula dos sedes (Madrid y Valencia).
# Cada sede tiene 3 celdas (antenas).
# Cada celda atiende usuarios (UE) numerados del 1 al 30.

# Tres slices 5G:
    # eMBB: Enhanced Mobile Broadband (alta velocidad).
    # URLLC: Ultra Reliable Low Latency (baja latencia, alta fiabilidad).
    # IoT: Internet of Things (bajo ancho de banda, muchos dispositivos).

data = []

for minute in range(minutes):
    timestamp = start + datetime.timedelta(minutes=minute)
    for site in sites:
        for cell in cells:
            slice_id = np.random.choice(slices, p=[0.6, 0.2, 0.2])
            for user in np.random.choice(users, size=10, replace=False):
                rsrp = np.random.normal(-85, 5)
                rsrq = np.random.normal(-10, 2)
                sinr = np.random.normal(20, 5)
                prb_util = np.clip(np.random.normal(60, 20), 0, 100)
                throughput_dl = max(0, np.random.normal(100 - (100 - prb_util)/2, 15))
                throughput_ul = max(0, throughput_dl * np.random.uniform(0.2, 0.4))
                latency = np.random.normal(10 + prb_util/20, 3)
                jitter = np.random.normal(latency/10, 1)
                pkt_loss = abs(np.random.normal(0.2 + (100 - sinr)/200, 0.1))
                ho_success = np.clip(99 - (100 - sinr)/5, 80, 99.9)
                energy = np.random.normal(35 + prb_util/5, 3)
                alarm_rate = np.clip((100 - ho_success)/1000, 0, 0.05)
                
                data.append({
                    'time': timestamp,
                    'site': site,
                    'cell_id': cell,
                    'user_id': user,
                    'slice_id': slice_id,
                    'rsrp_dbm': rsrp,
                    'rsrq_db': rsrq,
                    'sinr_db': sinr,
                    'throughput_dl_mbps': throughput_dl,
                    'throughput_ul_mbps': throughput_ul,
                    'latency_ms': latency,
                    'jitter_ms': jitter,
                    'packet_loss_pct': pkt_loss,
                    'prb_utilization_pct': prb_util,
                    'ho_success_rate': ho_success,
                    'energy_watts': energy,
                    'alarm_rate': alarm_rate
                })

df = pd.DataFrame(data)
df.to_csv("synthetic_5g_kpis.csv", index=False)

# Por cada minuto → cada sede → cada celda → elige un tipo de servicio (“slice”) con más probabilidad de ser eMBB (60%).
# Luego selecciona 10 usuarios activos aleatorios en ese minuto.
```
